```python
popularity_data = movie_data[['title' , 'rating']]
```

```python
popularity_data
```

#### calcualte the average ratings

```python
rating_mean_count = pd.DataFrame(popularity_data.groupby('title')['rating'].mean())
rating_mean_count
```

#### Calculate the count of ratings and add to the average ratings above

```python
rating_mean_count['rating_counts'] = pd.DataFrame(popularity_data.groupby('title')['rating'].count())
```

```python
rating_mean_count.describe()
```

```python
rating_mean_count.head()
```

```python
rating_mean_count['rating'] = round(rating_mean_count['rating'] , 1)
```

```python
rating_mean_count.head()
```

```python
rating_popularity =  rating_mean_count[(rating_mean_count['rating'] > 4)& (rating_mean_count['rating_counts'] > 15000)]
```

```python
rating_popularity
```

```python
rating_popularity = rating_popularity.sort_values(by='rating' , ascending=False)
rating_popularity
```
